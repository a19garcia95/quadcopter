```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('reward.txt')

plt.plot(df['episode'], df['total_reward'])
plt.xlabel('episode')
plt.ylabel('total reward')
plt.title('history of total reward')
_ = plt.ylim()
```

```python
plt.plot(df['episode'], df['x'], label='x')
plt.plot(df['episode'], df['y'], label='y')
plt.plot(df['episode'], df['z'], label='z')
plt.xlabel('episode')
plt.title('final position in each episode')
plt.legend()
_ = plt.ylim()
```

```python
plt.plot(df['episode'], df['x_velocity'], label='v_x')
plt.plot(df['episode'], df['y_velocity'], label='v_y')
plt.plot(df['episode'], df['z_velocity'], label='v_z')
plt.xlabel('episode')
plt.legend()
_ = plt.ylim()
```

```python
plt.plot(df['episode'], df['phi_velocity'], label='v_phi')
plt.plot(df['episode'], df['theta_velocity'], label='v_theta')
plt.plot(df['episode'], df['psi_velocity'], label='v_psi')
plt.xlabel('episode')
plt.legend()
_ = plt.ylim()
```
